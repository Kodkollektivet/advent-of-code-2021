```python
f = open("drive/MyDrive/advent/5/run.txt", "r")
lines=f.readlines()
print(lines)
```

```python
ventlines = []
for line in lines:
  line = line.strip()
  edges = line.split(' -> ')
  startCoordinates = {'x':int(edges[0].split(',')[0]), 'y':int(edges[0].split(',')[1])}
  stopCoordinates = {'x':int(edges[1].split(',')[0]), 'y':int(edges[1].split(',')[1])}
  ventline = {'start':startCoordinates, 'stop':stopCoordinates}

  ventlines.append(ventline)
print(ventlines)
```

```python
for ventline in ventlines:
  dots = []
  if ventline['start']['x'] == ventline['stop']['x']: # vertical:
    if ventline['start']['y'] < ventline['stop']['y']: # up
      for i in range (ventline['start']['y'], ventline['stop']['y']+1):
        dots.append({'x': ventline['start']['x'], 'y':i})
    
    if ventline['start']['y'] > ventline['stop']['y']: # down
      for i in range (ventline['stop']['y'], ventline['start']['y']+1):
        dots.append({'x': ventline['start']['x'], 'y':i})

  if ventline['start']['y'] == ventline['stop']['y']: # horizontal:
    if ventline['start']['x'] < ventline['stop']['x']: # rigth
      for i in range (ventline['start']['x'], ventline['stop']['x']+1):
        dots.append({'x': i, 'y' : ventline['start']['y']})

    if ventline['start']['x'] > ventline['stop']['x']: # left
      for i in range (ventline['stop']['x'], ventline['start']['x']+1):
        dots.append({'x': i, 'y' : ventline['start']['y']})

  
  ventline['dots'] = dots
  print(dots)
```

```python
for ventline in ventlines:
  print(ventline)
```

```python
overlaps = set(());

for i in range(0, len(ventlines)):
  for j in range(i+1,len(ventlines)):
    for dot in ventlines[i]['dots']:
      if dot in ventlines[j]['dots']:
        overlaps.add(f"{dot['x']}{dot['y']}")

print(len(overlaps))
```

```python
f = open("drive/MyDrive/advent/5/run.txt", "r")
lines=f.readlines()
print(lines)
```

```python
ventlines2 = []
for line in lines:
  line = line.strip()
  edges = line.split(' -> ')
  startCoordinates = {'x':int(edges[0].split(',')[0]), 'y':int(edges[0].split(',')[1])}
  stopCoordinates = {'x':int(edges[1].split(',')[0]), 'y':int(edges[1].split(',')[1])}
  ventline = {'start':startCoordinates, 'stop':stopCoordinates}

  ventlines2.append(ventline)
print(ventlines2)
```

```python
for ventline in ventlines2:
  dots = []
  ventline['dots'] = dots
  if ventline['start']['x'] == ventline['stop']['x']: # vertical:
    if ventline['start']['y'] < ventline['stop']['y']: # up
      for i in range (ventline['start']['y'], ventline['stop']['y']+1):
        dots.append({'x': ventline['start']['x'], 'y':i})
      continue
    
    if ventline['start']['y'] > ventline['stop']['y']: # down
      for i in range (ventline['stop']['y'], ventline['start']['y']+1):
        dots.append({'x': ventline['start']['x'], 'y':i})
      continue

  elif ventline['start']['y'] == ventline['stop']['y']: # horizontal:
    if ventline['start']['x'] < ventline['stop']['x']: # rigth
      for i in range (ventline['start']['x'], ventline['stop']['x']+1):
        dots.append({'x': i, 'y' : ventline['start']['y']})
      continue

    if ventline['start']['x'] > ventline['stop']['x']: # left
      for i in range (ventline['stop']['x'], ventline['start']['x']+1):
        dots.append({'x': i, 'y' : ventline['start']['y']})
      continue

  else: # diagonal:
    if ventline['start']['y'] < ventline['stop']['y'] and ventline['start']['x'] < ventline['stop']['x']: # up-right
      for i in range (0, (ventline['stop']['y'] - ventline['start']['y'])+1):
        dots.append({'x': ventline['start']['x'] + i, 'y':ventline['start']['y']+i})
      continue  

    if ventline['start']['y'] < ventline['stop']['y'] and ventline['start']['x'] > ventline['stop']['x']: # up-left
      for i in range (0, (ventline['stop']['y'] - ventline['start']['y'])+1):
        dots.append({'x': ventline['start']['x'] - i, 'y':ventline['start']['y']+i})
      continue
  
    if ventline['start']['y'] > ventline['stop']['y'] and ventline['start']['x'] < ventline['stop']['x']: # down-right
      for i in range (0, (ventline['stop']['x'] - ventline['start']['x'])+1):
        dots.append({'x': ventline['start']['x'] + i, 'y':ventline['start']['y']-i})
      continue

    if ventline['start']['y'] > ventline['stop']['y'] and ventline['start']['x'] > ventline['stop']['x']: # down-left
      for i in range (0, (ventline['start']['x'] - ventline['stop']['x'])+1):
        dots.append({'x': ventline['start']['x'] - i, 'y':ventline['start']['y']-i})
      continue

  
```

```python
overlaps2 = set(());

for i in range(0, len(ventlines2)):
  for j in range(i+1,len(ventlines2)):
    for dot in ventlines2[i]['dots']:
      if dot in ventlines2[j]['dots']:
        overlaps2.add(f"{dot['x']}{dot['y']}")

print(len(overlaps2))
```

```python
map = {}
count = 0
for ventline in ventlines2:
  for dot in ventline['dots']:
    if f"x{dot['x']}y{dot['y']}" in map.keys():
      map[f"{dot['x']}{dot['y']}"] = map[f"x{dot['x']}y{dot['y']}"] + 1
    else:
      map[f"x{dot['x']}y{dot['y']}"] = 1

for key in map:
  if map[key] > 1:
    count = count + 1

print(map)
```
