```python
f = open("drive/MyDrive/advent/4/boards.txt", "r")
boards=f.readlines()
print(boards)
```

```python
orderarr = order[0].split(',')
```

```python
orderarrnum = []
for num in orderarr:
  orderarrnum.append(int(num))

print(orderarrnum)
```

```python
boardsarr = []
board = {'numbers': [], 'rows': [], 'columns':[[],[],[],[],[]] }

for line in boards:
  if line == '\n':
    boardsarr.append(board)
    board = {'numbers': [], 'rows': [], 'columns':[[],[],[],[],[]] }
  else:
    
    line = line.strip()
    row = line.split(' ')
    rowNumb = []
    column = 0
    for numb in row:
      if numb == '':
        continue
      rowNumb.append(int(numb))
      board['columns'][column].append(int(numb))
      column = column + 1
      board['numbers'].append(int(numb))
    board['rows'].append(rowNumb)

if len(board['numbers']) != 0:
  boardsarr.append(board)
print(boardsarr)
```

```python
haveWinner = False
winner = None
for newnumb in orderarrnum:
  print('NEW NUMBER', newnumb)
  for board in boardsarr:
    for row in board['rows']:
      if newnumb in row:
        row.remove(newnumb)
        if len(row) == 0:
          haveWinner = True
    for column in board['columns']:
      if newnumb in column:
        column.remove(newnumb)
        if len(column) == 0:
          haveWinner = True
    if newnumb in board['numbers']:
      board['numbers'].remove(newnumb)
    print(board)

    if haveWinner:
      print('HAVE WINNER!', board, 'After', newnumb)
      winner = board
      break;
  if haveWinner:
    break;
```

```python
from functools import reduce

sum = reduce(lambda a,b: a+b, winner['numbers'])
print(sum * newnumb)
```

```python
boardsarr = []
board = {'numbers': [], 'rows': [], 'columns':[[],[],[],[],[]], 'status':'active' }

for line in boards:
  if line == '\n':
    boardsarr.append(board)
    board = {'numbers': [], 'rows': [], 'columns':[[],[],[],[],[]], 'status':'active' }
  else:
    
    line = line.strip()
    row = line.split(' ')
    rowNumb = []
    column = 0
    for numb in row:
      if numb == '':
        continue
      rowNumb.append(int(numb))
      board['columns'][column].append(int(numb))
      column = column + 1
      board['numbers'].append(int(numb))
    board['rows'].append(rowNumb)

if len(board['numbers']) != 0:
  boardsarr.append(board)
print(boardsarr)
```

```python

loserboard = None
for newnumb in orderarrnum:
  print('NEW NUMBER', newnumb)
  for board in boardsarr:
    haveWinner = False
    if board['status'] == 'removed':
      continue
    for row in board['rows']:
      if newnumb in row:
        row.remove(newnumb)
        if len(row) == 0:
          haveWinner = True
    for column in board['columns']:
      if newnumb in column:
        column.remove(newnumb)
        if len(column) == 0:
          haveWinner = True
    if newnumb in board['numbers']:
      if newnumb == 13:
        print('13')
      board['numbers'].remove(newnumb)
    print(board)

    if haveWinner:
      print('HAVE WINNER!', board, 'After', newnumb)
      loserboard = board
      board['status'] = 'removed'      

  stopGame = True
  for board in boardsarr:
    if board['status'] == 'active':
      stopGame = False

  if stopGame:
    break

print('LOSER', loserboard)
```

```python
sum = reduce(lambda a,b: a+b, loserboard['numbers'])
print(loserboard)
print(sum * newnumb)
```
