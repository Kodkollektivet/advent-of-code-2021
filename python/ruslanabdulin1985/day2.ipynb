```python
f = open("drive/MyDrive/advent/2/run.txt", "r")
lines=f.readlines()
print(lines)
```

```python
controls = []
DIRECTION = 'direction'
VALUE = 'value'
FORWARD = 'forward'
DOWN = 'down'
UP = 'up'


for line in lines:
  control = line.split(' ')
  controls.append({DIRECTION : control[0], VALUE : int(control[1])})

print(controls)
```

```python
depth = 0
distance = 0

for control in controls:
  if control[DIRECTION] == DOWN:
    depth = depth + control[VALUE]
  
  if control[DIRECTION] == UP:
    depth = depth - control[VALUE]

  if control[DIRECTION] == FORWARD:
    distance = distance + control[VALUE]

result = depth*distance
print(result)
```

```python
depth = 0
distance = 0
aim = 0

for control in controls:
  if control[DIRECTION] == DOWN:
    aim = aim + control[VALUE]
  
  if control[DIRECTION] == UP:
    aim = aim - control[VALUE]

  if control[DIRECTION] == FORWARD:
    distance = distance + control[VALUE]
    depth = depth + (control[VALUE] * aim)


result = depth*distance
print(result)
```
