```python
f = open("drive/MyDrive/advent/3/run.txt", "r")
lines=f.readlines()
print(lines)
```

```python
gamma = ''
epsilon = ''

for line in lines:
  line.strip()

print(lines)

length = len(lines[0])-1

for i in range (length):

  zeroes = 0
  ones = 0
  for line in lines:
    firstBit = int(line[i])
    if firstBit == 0:
      zeroes = zeroes + 1

    if firstBit == 1:
      ones = ones + 1

  # print(firstBit)

  bit = '0' if zeroes > ones else '1'
  gamma = gamma + bit
  bit = '1' if zeroes > ones else '0'
  epsilon = epsilon + bit

print ('gamma', int(gamma, 2))
print ('epsilon', int(epsilon, 2))
result = int(gamma, 2) * int(epsilon, 2)
print('result', result)
```

```python
pos =0

lines1 = lines.copy()

while (len(lines1)>1):
  ones = list(filter(lambda line: line[pos]=='1', lines1))
  zeroes = list(filter(lambda line: line[pos]=='0', lines1))
  lines1 = ones if len(ones)>=len(zeroes) else zeroes
  pos = pos + 1

print (lines1)
```
