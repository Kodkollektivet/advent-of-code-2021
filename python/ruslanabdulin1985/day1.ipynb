```python
f = open("drive/MyDrive/advent/1/run.txt", "r")
lines=f.readlines()
print(lines)
```

```python

numbers = []
previous = 0
increasing = 0
for line in lines:
  number = int(line)
  if number > previous and previous != 0:
    increasing = increasing + 1
  numbers.append(number)
  previous = number

print(increasing)
print (numbers)
print (len(numbers))
```

```python
first = -1
second = -1
third = -1
sums = []

for number in numbers:
  if first==-1:
    first=number
    continue
  
  if second==-1:
    second=number
    continue
  
  if third == -1:
    third = number

  sums.append(first+second+third)
  first = second
  second = third
  third = -1 

print(sums)


```

```python
increasing = 0
previous = 0

print(len(sums))

for sum in sums:
  if previous == 0:
    previous = sum
    continue
  if sum > previous:
    increasing =  increasing + 1 
  previous = sum

print(increasing)
    
```
